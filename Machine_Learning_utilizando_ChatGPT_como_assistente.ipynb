#Normalizando vari√°veis num√©ricas
##Prompt ü§ñ

Nas colunas "Age", "Income", "Family", "CCAvg", "Mortgage" temos diferentes grandezas de valores.
√â necess√°rio mudar as escalas dos valores para criar os modelos de Machine Learning?

**Resposta do ChatGPT:**

Existem v√°rias t√©cnicas comuns para normalizar ou padronizar os valores em colunas num√©ricas. Alguns exemplos incluem:

*   Min-Max Scaling (Normaliza√ß√£o): Essa t√©cnica dimensiona os valores em um intervalo espec√≠fico, geralmente entre 0 e 1. A f√≥rmula para aplicar a normaliza√ß√£o min-max √©: >X_scaled = (X - X_min) / (X_max - X_min). Onde X representa o valor original, X_min √© o valor m√≠nimo na coluna e X_max √© o valor m√°ximo na coluna.

*   Z-Score Standardization (Padroniza√ß√£o): Essa t√©cnica transforma os valores para que tenham uma m√©dia de zero e um desvio padr√£o de um. A f√≥rmula para aplicar a padroniza√ß√£o z-score √©: X_scaled = (X - mean) / std. Onde X representa o valor original, mean √© a m√©dia da coluna e std √© o desvio padr√£o da coluna.

*   Robust Scaling (Escalonamento Robusto): Essa t√©cnica √© √∫til quando os dados possuem outliers ou est√£o distribu√≠dos de forma n√£o normal. Ela utiliza a mediana e o intervalo interquartil para dimensionar os valores. A f√≥rmula para aplicar o escalonamento robusto √©: X_scaled = (X - median) / (Q3 - Q1). Onde X representa o valor original, median √© a mediana da coluna, Q1 √© o primeiro quartil (25%) e Q3 √© o terceiro quartil (75%).

Essas s√£o apenas algumas t√©cnicas comuns de normaliza√ß√£o ou padroniza√ß√£o. A escolha da t√©cnica adequada depende do seu conjunto de dados, da distribui√ß√£o dos valores e dos requisitos espec√≠ficos do seu modelo de Machine Learning. √â importante experimentar diferentes t√©cnicas e observar o impacto nos resultados do seu modelo. A biblioteca scikit-learn em Python oferece implementa√ß√µes dessas t√©cnicas e outras transforma√ß√µes √∫teis para normaliza√ß√£o de dados.

Aplicar a t√©cnica de one-hot na coluna "Education" com a fun√ß√£o "get_dummies()" da biblioteca pandas.
Devolver o DataFrame pr√©-processado.

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

def preprocess_data(df):
    # Remover as colunas "ID", "ZIP Code" e "Experience"
    df = df.drop(columns=["ID", "ZIP Code", "Experience"])

    # Retirar outliers da coluna "Mortgage" usando zscore
    z_scores = zscore(df["Mortgage"])
    df = df[(np.abs(z_scores) < 3)]

    # Multiplicar os valores da coluna "CCAvg" por 12
    df["CCAvg"] = df["CCAvg"] * 12

    # Normalizar as colunas "Age", "Income", "Family", "CCAvg" e "Mortgage" usando MinMaxScaler
    scaler = MinMaxScaler()
    columns_to_normalize = ["Age", "Income", "Family", "CCAvg", "Mortgage"]
    df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

    # Aplicar a t√©cnica de one-hot encoding na coluna "Education" usando get_dummies
    df = pd.get_dummies(df, columns=["Education"])

    return df
```

 # Algoritmo Regress√£o log√≠stica

 O "Logistic Regression Classifier" √© um algoritmo de aprendizado de m√°quina amplamente utilizado para problemas de classifica√ß√£o bin√°ria, onde o objetivo √© prever a probabilidade de um exemplo pertencer a uma das duas classes poss√≠veis.

 A ideia b√°sica por tr√°s da logistic regression √© modelar a rela√ß√£o entre as vari√°veis de entrada (ou recursos) e a probabilidade de pertencer a uma classe espec√≠fica usando uma fun√ß√£o log√≠stica, tamb√©m conhecida como fun√ß√£o sigm√≥ide. Essa fun√ß√£o tem a propriedade de transformar valores reais em um intervalo entre 0 e 1, o que √© ideal para representar probabilidades.

O algoritmo de logistic regression calcula um conjunto de coeficientes ou pesos para cada vari√°vel de entrada, que s√£o usados para ponderar a import√¢ncia de cada vari√°vel na previs√£o da classe. Esses coeficientes s√£o estimados usando uma t√©cnica chamada de "m√°xima verossimilhan√ßa" ou outros m√©todos de otimiza√ß√£o, como o gradiente descendente.

Durante o treinamento, o algoritmo ajusta os coeficientes para minimizar a diferen√ßa entre as probabilidades previstas e as classes reais dos exemplos de treinamento. Uma vez que os coeficientes s√£o estimados, eles podem ser usados para fazer previs√µes em novos exemplos.

Para fazer uma previs√£o, o algoritmo calcula a soma ponderada das vari√°veis de entrada, usando os coeficientes aprendidos. Essa soma √© ent√£o passada pela fun√ß√£o sigmoide, que retorna a probabilidade estimada de pertencer √† classe positiva. Tipicamente, um limiar de decis√£o √© escolhido (geralmente 0,5) para classificar exemplos com base nessa probabilidade. Se a probabilidade estimada for maior que o limiar, o exemplo √© classificado como pertencente √† classe positiva; caso contr√°rio, √© classificado como pertencente √† classe negativa.

A logistic regression pode lidar com m√∫ltiplas vari√°veis de entrada e tamb√©m pode ser estendida para problemas de classifica√ß√£o multiclasse usando t√©cnicas como a regress√£o log√≠stica multinomial ou a regress√£o log√≠stica ordinal.

√â importante ressaltar que a logistic regression tem suposi√ß√µes subjacentes, como a rela√ß√£o linear entre as vari√°veis de entrada e a log-odds da classe positiva. Al√©m disso, a logistic regression pode ser afetada por problemas como multicolinearidade e outliers, que podem influenciar a estimativa dos coeficientes.

Em resumo, o "Logistic Regression Classifier" √© um algoritmo de aprendizado de m√°quina que usa a fun√ß√£o sigm√≥ide para modelar a rela√ß√£o entre as vari√°veis de entrada e a probabilidade de pertencer a uma classe espec√≠fica. Ele √© treinado ajustando os coeficientes para minimizar a diferen√ßa entre as probabilidades previstas e as classes reais, e pode ser usado para fazer previs√µes em novos exemplos, classificando-os com base nas probabilidades estimadas.

# Algoritmo √Årvore de decis√£o
O algoritmo "Decision Tree Classifier" (classificador de √°rvore de decis√£o) √© um m√©todo de aprendizado de m√°quina que utiliza √°rvores de decis√£o para resolver problemas de classifica√ß√£o. Ele constr√≥i uma estrutura em forma de √°rvore, onde cada n√≥ interno representa uma condi√ß√£o ou atributo, cada ramo representa uma resposta a essa condi√ß√£o, e cada folha representa uma classe ou r√≥tulo de sa√≠da.

A constru√ß√£o da √°rvore de decis√£o ocorre em etapas. Inicialmente, o algoritmo seleciona o atributo que melhor separa os exemplos de treinamento em rela√ß√£o √†s classes desejadas. Isso √© feito calculando uma m√©trica de impureza, como a entropia ou o √≠ndice de Gini, que mede a heterogeneidade dos exemplos em cada ramo da √°rvore.

Uma vez que um atributo √© escolhido, a √°rvore √© dividida em ramos com base nos poss√≠veis valores desse atributo. Cada ramo representa uma resposta a essa condi√ß√£o e leva a um novo n√≥ interno ou a uma folha. Esse processo √© repetido recursivamente para cada ramo, levando √† cria√ß√£o de sub√°rvores.

A constru√ß√£o da √°rvore continua at√© que uma condi√ß√£o de parada seja atingida. Isso pode ocorrer quando todos os exemplos em um ramo pertencem √† mesma classe, quando n√£o h√° mais atributos dispon√≠veis para dividir, ou quando uma determinada profundidade m√°xima da √°rvore √© alcan√ßada.

Para fazer uma previs√£o usando a √°rvore de decis√£o treinada, um exemplo √© passado pela estrutura da √°rvore, seguindo as condi√ß√µes em cada n√≥. Finalmente, o exemplo alcan√ßa uma folha, que representa a classe prevista para esse exemplo.

A √°rvore de decis√£o pode lidar com atributos categ√≥ricos e num√©ricos, al√©m de lidar com dados ausentes. Quando os atributos s√£o num√©ricos, o algoritmo pode usar estrat√©gias de divis√£o, como "maior que" ou "menor que", para realizar as divis√µes nos ramos.

Uma das vantagens da √°rvore de decis√£o √© sua interpretabilidade, pois √© poss√≠vel visualizar a estrutura da √°rvore e entender as decis√µes tomadas em cada n√≥. Al√©m disso, as √°rvores de decis√£o podem lidar com conjuntos de dados grandes e complexos.

No entanto, as √°rvores de decis√£o tamb√©m podem ser propensas a superajuste (overfitting), ou seja, se adaptarem muito aos dados de treinamento e n√£o generalizarem bem para novos exemplos. Para mitigar esse problema, t√©cnicas como a poda da √°rvore, a defini√ß√£o de profundidade m√°xima ou o uso de ensemble methods, como o Random Forest e o Gradient Boosting, podem ser aplicadas.

Em resumo, o algoritmo "Decision Tree Classifier" constr√≥i uma estrutura em forma de √°rvore para resolver problemas de classifica√ß√£o. Ele divide os exemplos de treinamento com base nos atributos mais informativos e utiliza a estrutura da √°rvore para fazer previs√µes em novos exemplos, percorrendo os ramos at√© alcan√ßar as folhas.

##Matriz de confus√£o

Uma matriz de confus√£o, tamb√©m conhecida como tabela de conting√™ncia, √© uma ferramenta comumente usada em machine learning para avaliar o desempenho de um modelo de classifica√ß√£o. Ela fornece uma representa√ß√£o visual da qualidade das previs√µes feitas pelo modelo, comparando os resultados esperados com os resultados obtidos.

A matriz de confus√£o √© constru√≠da com base nos quatro poss√≠veis resultados de uma classifica√ß√£o bin√°ria:

Verdadeiro Positivo (True Positive - TP): Refere-se aos casos em que o modelo previu corretamente a classe positiva (verdadeira) de uma amostra. Por exemplo, se um modelo de detec√ß√£o de spam classificar corretamente um e-mail como spam, ser√° considerado um verdadeiro positivo.

Falso Positivo (False Positive - FP): Ocorre quando o modelo prev√™ incorretamente a classe positiva (verdadeira) para uma amostra que pertence √† classe negativa (falsa). Nesse caso, o modelo de detec√ß√£o de spam rotula um e-mail leg√≠timo como spam.

Verdadeiro Negativo (True Negative - TN): Indica os casos em que o modelo prev√™ corretamente a classe negativa (falsa) para uma amostra. Seria quando o e-mail n√£o √© spam e o modelo o identifica corretamente como tal.

Falso Negativo (False Negative - FN): Refere-se aos casos em que o modelo prev√™ incorretamente a classe negativa (falsa) para uma amostra que pertence √† classe positiva (verdadeira). Um falso negativo ocorre quando o modelo falha em identificar corretamente um e-mail como spam, quando, na verdade, ele √© um spam. Por exemplo, se um spam chega √† caixa de entrada e o modelo n√£o o marca como spam.

A matriz de confus√£o √© uma tabela 2x2 que organiza esses resultados da seguinte maneira:

Classe Positiva	    | Classe Negativa

Positivo (Previsto)	| TP

Negativo (Previsto)	| FN

Cada c√©lula da matriz representa uma contagem do n√∫mero de inst√¢ncias que caem em uma determinada categoria. A diagonal principal da matriz (da esquerda superior para a direita inferior) cont√©m os verdadeiros positivos e verdadeiros negativos, enquanto as c√©lulas fora da diagonal principal cont√©m os falsos positivos e falsos negativos.

A matriz de confus√£o fornece uma vis√£o geral do desempenho do modelo em termos de sua capacidade de fazer previs√µes corretas e identificar corretamente as classes positivas e negativas. Com base nessa matriz, v√°rias m√©tricas de avalia√ß√£o podem ser calculadas, incluindo a acur√°cia, precis√£o, recall e F1-score, que fornecem uma compreens√£o mais aprofundada do desempenho do modelo.

## A matem√°tica por tr√°s das m√©tricas

A fun√ß√£o classification_report da biblioteca sklearn √© usada para calcular v√°rias m√©tricas de avalia√ß√£o para um problema de classifica√ß√£o. Essas m√©tricas incluem precis√£o (precision), recall (recall), F1-score (f1-score) e suporte (support) para cada classe.

Aqui est√° uma explica√ß√£o detalhada de como essas m√©tricas s√£o calculadas:

**Precis√£o (Precision)**: A precis√£o √© calculada como a raz√£o entre o n√∫mero de
verdadeiros positivos (VP) e a soma dos verdadeiros positivos e falsos positivos (FP). Em termos matem√°ticos, a f√≥rmula da precis√£o √©:

*   Precis√£o = VP / (VP + FP)

A precis√£o mede a propor√ß√£o de exemplos positivos classificados corretamente em rela√ß√£o a todos os exemplos classificados como positivos. Ela fornece uma indica√ß√£o da confiabilidade das previs√µes positivas.

**Recall (Recall**): O recall, tamb√©m conhecido como taxa de verdadeiros positivos, √© calculado como a raz√£o entre o n√∫mero de verdadeiros positivos (VP) e a soma dos verdadeiros positivos e falsos negativos (FN). A f√≥rmula do recall √©:

*   Recall = VP / (VP + FN)

O recall representa a propor√ß√£o de exemplos positivos corretamente classificados em rela√ß√£o a todos os exemplos que deveriam ter sido classificados como positivos. Ele mede a capacidade do modelo em encontrar todos os exemplos positivos.

**F1-score (F1-Score)**: O F1-score √© uma m√©dia harm√¥nica ponderada de precis√£o e recall, combinando essas duas m√©tricas em uma √∫nica medida. √â calculado usando a f√≥rmula:

*   F1-score = 2 * (Precis√£o * Recall) / (Precis√£o + Recall)

O F1-score fornece uma medida balanceada da precis√£o e do recall, levando em considera√ß√£o tanto os falsos positivos quanto os falsos negativos. √â √∫til quando h√° um desequil√≠brio significativo entre as classes.

**Suporte (Support)**: O suporte √© simplesmente a contagem de exemplos de cada classe presente no conjunto de dados de teste. Ele mostra quantas inst√¢ncias pertencem a cada classe.

Essas m√©tricas s√£o calculadas para cada classe individualmente e tamb√©m podem ser calculadas para uma m√©dia ponderada (m√©dia micro) ou n√£o ponderada (m√©dia macro) de todas as classes.

O classification_report da biblioteca sklearn gera uma tabela que apresenta essas m√©tricas para cada classe e tamb√©m as m√©dias ponderadas e n√£o ponderadas. Essas m√©tricas nos fornecem informa√ß√µes valiosas sobre o desempenho do modelo em termos de precis√£o, recall e F1-score para cada classe, ajudando-nos a avaliar a qualidade das previs√µes.

Esses m√©todos de valida√ß√£o cruzada permitem uma avalia√ß√£o mais robusta do desempenho do modelo, ajudando a identificar problemas de overfitting, subajuste ou outros problemas de generaliza√ß√£o.

## Como funciona a valida√ß√£o cruzada
A valida√ß√£o cruzada (cross-validation) √© uma t√©cnica utilizada em machine learning para avaliar o desempenho de um modelo de forma mais robusta e confi√°vel. Em vez de simplesmente dividir o conjunto de dados em apenas um conjunto de treinamento e um conjunto de teste, a valida√ß√£o cruzada envolve dividir o conjunto de dados em v√°rias partes (chamadas de folds) e executar o processo de treinamento e teste em cada uma dessas partes. Essa divis√£o em conjuntos de treinamento e teste pode ser feita de v√°rias maneiras, sendo o K-Fold Cross-Validation e o Stratified K-Fold Cross-Validation duas abordagens comumente utilizadas.

No K-Fold Cross-Validation, o conjunto de dados √© dividido em K partes iguais (chamadas de folds ou dobras). O modelo √© treinado K vezes, cada vez usando K-1 folds como conjunto de treinamento e 1 fold como conjunto de teste. O desempenho do modelo √© calculado como a m√©dia dos resultados obtidos em cada itera√ß√£o.

J√° no Stratified K-Fold Cross-Validation, a divis√£o dos dados em folds √© feita de forma a preservar a propor√ß√£o de classes em cada fold. Isso √© especialmente √∫til quando se lida com problemas de classifica√ß√£o em que as classes n√£o est√£o balanceadas no conjunto de dados. O objetivo √© garantir que cada fold represente adequadamente a distribui√ß√£o de classes do conjunto de dados original.

Em resumo, enquanto o K-Fold Cross-Validation divide o conjunto de dados em partes iguais, o Stratified K-Fold Cross-Validation leva em considera√ß√£o a distribui√ß√£o de classes ao realizar essa divis√£o. Ambas as abordagens s√£o √∫teis para avaliar o desempenho de modelos de machine learning e escolher os melhores hiperpar√¢metros, mas o Stratified K-Fold Cross-Validation √© mais recomendado quando se lida com classes desbalanceadas.

* **Combina√ß√£o de Undersampling e Oversampling**: Tamb√©m √© poss√≠vel combinar as t√©cnicas de undersampling e oversampling para obter um balanceamento mais efetivo. A biblioteca imbalanced-learn oferece m√©todos que combinam essas abordagens, como SMOTEENN e SMOTETomek.

* **Penaliza√ß√£o na fun√ß√£o de custo**: Alguns algoritmos de machine learning possuem par√¢metros que permitem penalizar de forma diferente os erros nas classes majorit√°ria e minorit√°ria durante o treinamento. ## Criar uma fun√ß√£o que realize a valida√ß√£o cruzada
O c√≥digo abaixo cria uma fun√ß√£o chamada roda_modelo_cv, que pede como argumento o nome do modelo que deve ser previamente instanciado, o n√∫mero de splits e o n√∫mero de repeti√ß√µes do RepeatedStratifiedKFold:

```python
def roda_modelo_cv(modelo, n_splits, n_repeats):

  cv = RepeatedStratifiedKFold(n_splits = n_splits, n_repeats=n_repeats, random_state=42)
  resultados=cross_validate(modelo, X_train_balanced, y_train_balanced, cv=cv, scoring='roc_auc', return_train_score=True)

  auc_medio_treino = np.mean(resultados['train_score']) #m√©dia do AUC do treino
  auc_medio_teste = np.mean(resultados['test_score']) #m√©dia do AUC do teste

  auc_std_treino = np.std(resultados['train_score']) #desvio-padr√£o do treino
  auc_std_teste = np.std(resultados['test_score']) #desvio-padr√£o do teste

  print(f"Modelo {modelo}")
  print(f"AUC m√©dio do treino {auc_medio_teste:.4f}/ Desvio-padr√£o treino {auc_std_teste:.4f}")
  print(f"AUC m√©dio do Teste {auc_medio_teste:.4f}/ Desvio-padr√£o teste {auc_std_teste:.4f}")
```
